```python
import pandas as pd
#import matplotlib
# Read in the data/HR_Employee_Attrition
df = pd.read_csv('data/HR_Employee_Attrition.csv')

# Print the first 5 rows of the dataframe
df.head()
```

```python
import matplotlib.pyplot as plt

#top 10 highest year in the company

top_10_longest_employeeworker = df.sort_values(by="YearsAtCompany", ascending=False)

#create a correct label for y-axis instead of 1600, 1400 etc


#create a visualization
plt.figure(figsize=(10,5))
plt.bar(top_10_longest_employeeworker['YearsAtCompany'].head(10), top_10_longest_employeeworker['EmployeeNo'].head(10), color='blue')
plt.xlabel("YearsAtCompany")
plt.ylabel("EmployeeNumber")


plt.title("Top 10 longest employee in the company")
plt.show()




```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame to include only employees with <= 5 years and attrition
filtered_df = df[(df['YearsAtCompany'] <= 5) & (df['Attrition'] == 'Yes')]

# Plot the scatter plot for filtered data
filtered_df.plot.scatter(x='YearsAtCompany', y='Attrition')

# Set labels for the axes
plt.xlabel('Years at Company')
plt.ylabel('Attrition')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you already have a DataFrame named 'df'

# Sorting the DataFrame by "PerformanceRating" in descending order to get the top performers
top_5_performers = df.sort_values(by="PerformanceRating", ascending=True).head()

# Creating a visualization
plt.figure(figsize=(10, 5))
plt.bar(top_5_performers['EmployeeNo'], top_5_performers['PerformanceRating'], color='blue')
plt.xlabel("EmployeeNumber")
plt.ylabel("PerformanceRating")
plt.title('Top Performing Employees by Performance Rating')
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


```
