```python
import pandas as pd
#import matplotlib
# Read in the data/HR_Employee_Attrition
df = pd.read_csv('data/HR_Employee_Attrition.csv')

# Print the first 5 rows of the dataframe
df.head()
```

```python
import matplotlib.pyplot as plt

# Plot the distribution of the YearsAtCompany atleast 5 years column
df.plot.scatter(x='Attrition', y='YearsAtCompany')

plt.show()

```

```python
import matplotlib.pyplot as plt

# Filter the DataFrame to include only employees with <= 5 years and attrition
filtered_df = df[(df['YearsAtCompany'] <= 5) & (df['Attrition'] == 'Yes')]

# Plot the scatter plot for filtered data
filtered_df.plot.scatter(x='YearsAtCompany', y='Attrition')

# Set labels for the axes
plt.xlabel('Years at Company')
plt.ylabel('Attrition')

# Show the plot
plt.show()
```
