```python
#to make it permanent, use inplace = True
data.rename(columns = {"SalaryPerWeek" : "SPW"}, inplace=True)
data.head()
```

```python
import matplotlib.pyplot as plt

top_10_salaries = data.sort_values(by="Salary", ascending=False).head(10)

# Create a bar chart to visualize the top 10 salaries
plt.figure(figsize=(12, 6))
plt.bar(top_10_salaries["Name"], top_10_salaries["Salary"])
plt.xlabel("Player Name")
plt.ylabel("Salary")
plt.title("Top 10 Highest Salaries in the NBA")
plt.xticks(rotation=45, ha="right")


# Display the chart
plt.tight_layout()
plt.show()
```

```python
#top ten lowest salary in NBA data
top_10_lowest_salaries = data.sort_values(by="Salary", ascending=True).head(10)

# Create a bar chart to visualize the top 10 salaries
plt.figure(figsize=(12, 6))
plt.bar(top_10_lowest_salaries["Name"], top_10_lowest_salaries["Salary"])
plt.xlabel("Player Name")
plt.ylabel("Salary")
plt.title("Top 10 Lowest Salaries in the NBA")
plt.xticks(rotation=45, ha="right")
plt.show()
```
